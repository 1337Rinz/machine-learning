```python
#import lib
import pandas as pd
```

```python
# import dataset
df = pd.read_csv("https://raw.githubusercontent.com/RinzCSGO/machine-learning/main/data/ml-basics/salaries.csv")
df.head()
```

```python
# drop columns (Biến phụ thuộc)
inputs = df.drop('salary_more_then_100k',axis='columns')
```

```python
# check dataset
inputs
```

```python
taget = df['salary_more_then_100k']
#check taget
taget
```

```python
# fit transform (chữ thành số)
from sklearn.preprocessing import LabelEncoder
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()
```

```python
inputs['company_n'] = le_company.fit_transform(inputs['company'])
inputs['job_n'] = le_job.fit_transform(inputs['job'])
inputs['degree_n'] = le_degree.fit_transform(inputs['degree'])
inputs
```

```python
#drop cột chữ
inputs_n = inputs.drop(['company','job','degree'],axis='columns')
inputs_n
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_n, taget)
```

```python
# score check (0.0 -- 1.0)
model.score(inputs_n, taget)
```

Is salary of Google, Computer Engineer, Bachelors degree > 100k


```python
#predict 01
model.predict([[2,1,0]])
```

Is salary of Google, Computer Engineer, Masters degree > 100k


```python
#predict 02
model.predict([[2,1,1]])
```
